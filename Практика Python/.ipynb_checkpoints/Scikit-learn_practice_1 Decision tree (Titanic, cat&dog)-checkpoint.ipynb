```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
data = pd.DataFrame({'X_1': [1, 1, 1, 0, 0, 0, 0, 1], 
                     'X_2': [0, 0, 0, 1, 0, 0, 0, 1], 
                     'Y': [1, 1, 1, 1, 0, 0, 0, 0]})
data
```

```python
clf = tree.DecisionTreeClassifier(criterion = "entropy")
```

```python
X_features = data[['X_1', 'X_2']]
Y_res = data.Y
```

```python
# обучение дерева решений
clf.fit(X_features, Y_res)
```

```python
plt.figure(figsize=(7, 6))
tree.plot_tree(clf, feature_names=list(X_features),
               class_names=['Negative', 'Positive'],
               filled=True);
```

Цель дерева решений — снизить энтропию.

**Энтропия** (мера неопределенности). Энтропия — это то, как много информации вам не известно о системе.  
  
E = 1 максимальная степень неопределенности данных -> по данным события равновероятны  

**IG (information gain)** — критерий прироста информации.  
N — количество наблюдений, n1 и n2 — количество наблюдений в первой и второй ветке деления по признаку соответственно. Е1 и Е2 — энтропия в первой и второй ветке соответственно.  

Для нашего примера:
   


```python
def entropy(p):
    return (-p*np.log2(p)-(1-p)*np.log2(1-p))

def IG(Entrp, n1, n2, entrp1, entrp2):
    return Entrp - n1/N * entrp1 - n2/N * entrp2
```

### Overfitting! (переобучение)
**Переобученная модель — модель, погрязшая в особенностях тренировочного датасэта, вместо вникания в общие закономерности** 

Можно ограничить глубину дерева.  
Покажем дереву два набора данных с помощью train_test_split, которая разбивает данные на тренировочные и тестовые (33% от всех данных)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# предсказывает долю правильных ответов
clf.score(X, y)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

На тестовом датасете модель выделяет только 77% правильных ответов. Возможная проблема — переобучение.  
Попробуем ограничить глубину дерева.

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

Поиск оптимального max_depth (при котором train_score близок к test_score и довольно высок сам по себе)

```python
max_depth_values = np.arange(1, 100)
train_val = []
test_val = []
for max_depth in max_depth_values:
    clf_try = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf_try.fit(X_train, y_train)
    train_val.append(clf_try.score(X_train, y_train))
    test_val.append(clf_try.score(X_test, y_test))
```

```python
ax = sns.lineplot(x=max_depth_values, y=train_val, label='train')
ax = sns.lineplot(x=max_depth_values, y=test_val, label='test')
```

```python
plt.figure(figsize=(12, 7))
tree.plot_tree(clf, feature_names=list(X_train),
               class_names=['Died', 'Survived'],
               filled=True);
```

### Но! 
Проблема переобучения еще не решена, т.к. злоупотребляли сравнением train и test, используя один и тот же test для "подгона" score.
### Кросс-валидация — один из способов решения
Валидирующее множество: на нём после прогона train подгоняем параметры. И только в финале сравниваем с множеством test, которое модель "видит" впервые. Score на тестовых данных — самый важный показатель.  
*Train dataset* используется для тренировки модели, *validation* — для её проверки, *test* датасет выступает как ещё один уровень проверки, которая происходит на не показанных модели ранее данных.


```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4)
```

```python
cross_val_score(clf, X_train, y_train, cv=5)
# cv - на сколько fold хотим разбить данные
# возвращает точности классификатора, разбив данные на 5 кусков
# обучился на всех, кроме первого, потом на всех, кроме второго, и т.д.
```

```python
max_depth_values = np.arange(1, 100)
train_val = []
cross_val = []
for max_depth in max_depth_values:
    clf_try = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf_try.fit(X_train, y_train)
    
    train_val.append(clf_try.score(X_train, y_train))
    
    mean_cross_val_score = cross_val_score(clf_try, X_train, y_train, cv=5).mean()
    cross_val.append(mean_cross_val_score)
```

```python
ax = sns.lineplot(x=max_depth_values, y=train_val, label='train')
ax = sns.lineplot(x=max_depth_values, y=cross_val, label='cross_val')
```

```python
cross_val = np.array(cross_val)
np.where(cross_val == cross_val.max())[0] + 1
# max_depth, при котором cross_val_score максимальный
```

```python
best_clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=10)
best_clf.fit(X_train, y_train)
```

```python
best_clf.score(X_test, y_test)
# финальное значение
```

## Тренировка на данных кошки vs собаки. Предсказание

```python
catdog = pd.read_csv('dogs_n_cats.csv')
```

```python
catdog.head(3)
```

```python
X_cd = catdog.drop('Вид', axis=1)
y_cd = catdog['Вид']
```

```python
max_depth_values = np.arange(1, 100)
train_val_cd = []
cross_val_cd = []
for max_depth in max_depth_values:
    clf_try_cd = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf_try_cd.fit(X_cd, y_cd)
    
    train_val_cd.append(clf_try_cd.score(X_cd, y_cd))
    
    mean_cross_val_score_cd = cross_val_score(clf_try_cd, X_cd, y_cd, cv=5).mean()
    cross_val_cd.append(mean_cross_val_score_cd)
```

```python
ax = sns.lineplot(x=max_depth_values, y=train_val_cd, label='train')
ax = sns.lineplot(x=max_depth_values, y=cross_val_cd, label='cross_val')
```

```python
best_clf_cd = tree.DecisionTreeClassifier(criterion='entropy', max_depth=1)
best_clf_cd.fit(X_cd, y_cd)
```

```python
test_cd = pd.read_json('test_catdog.txt')
```

```python
result = best_clf_cd.predict(test_cd)
```

```python
result = pd.DataFrame(result)
result.value_counts()
```
