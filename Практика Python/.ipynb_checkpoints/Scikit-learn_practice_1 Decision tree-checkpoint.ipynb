```python
from sklearn import tree
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.DataFrame({'X_1': [1, 1, 1, 0, 0, 0, 0, 1], 
                     'X_2': [0, 0, 0, 1, 0, 0, 0, 1], 
                     'Y': [1, 1, 1, 1, 0, 0, 0, 0]})
data
```

```python
clf = tree.DecisionTreeClassifier(criterion = "entropy")
```

```python
X_features = data[['X_1', 'X_2']]
Y_res = data.Y
```

```python
# обучение дерева решений
clf.fit(X_features, Y_res)
```

```python
plt.figure(figsize=(7, 6))
tree.plot_tree(clf, feature_names=list(X_features),
               class_names=['Negative', 'Positive'],
               filled=True);
```

**Энтропия** (мера неопределенности). Энтропия — это то, как много информации вам не известно о системе.  
  
E = 1 максимальная степень неопределенности данных -> по данным события равновероятны  

**IG (information gain)** — критерий прироста информации.  
N — количество наблюдений, n1 и n2 — количество наблюдений в первой и второй ветке деления по признаку соответственно. Е1 и Е2 — энтропия в первой и второй ветке соответственно.  

Для нашего примера:
   


```python
def entropy(p):
    return (-p*np.log2(p)-(1-p)*np.log2(1-p))

def IG(Entrp, n1, n2, entrp1, entrp2):
    return Entrp - n1/N * entrp1 - n2/N * entrp2
```
